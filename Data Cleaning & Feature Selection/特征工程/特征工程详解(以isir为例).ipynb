### 数据预处理


#### 2.1 无量纲化

无量纲化使不同规格的数据转换到同一规格。常见的无量纲化方法有标准化和区间缩放法。

标准化的前提是特征值服从正态分布，标准化后，其转换成标准正态分布。

区间缩放法利用了边界值信息，将特征的取值区间缩放到某个特点的范围，例如[0, 1]等。

```python
# 标准化
from sklearn.preprocessing import StandardScaler
StandardScaler().fit_transform(iris.data)
```
