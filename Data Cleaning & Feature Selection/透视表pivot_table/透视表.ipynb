```python
import pandas as pd
import numpy as np
```

```python
L = "L:/基本语法/数据清洗特征选择/透视表pivot_table/James_Harden.txt"
f = open(L,encoding="UTF-8")
# res = pd.read_csv(f)
df = pd.read_csv(f)
df
```

#### 查看哈登对阵每个队伍的得分(各项取mean)

```python
pd.pivot_table(df,index=[u'对手'])
```

#### 查看对阵同一<b>对手<b> 在主客场的数据 (二重index)

```python
pd.pivot_table(df,index=[u'对手',u'主客场'])
```

```python
# 交换顺序
pd.pivot_table(df,index=[u'主客场',u'对手'])
```

```python
### 查看不同主客场的得分、篮板、助攻
```

```python
pd.pivot_table(df,index=[u'主客场',u'胜负'],values=[u'得分',u'助攻',u'篮板'])
```

### Aggfunc默认mean

##### 在主客场和不同胜负情况下的总得分、总篮板、总助攻

```python
pd.pivot_table(df,index=[u'主客场',u'胜负'],values=[u'得分',u'助攻',u'篮板'],aggfunc=[np.sum,np.mean])
```

# columns

```python
# fill_value 填充空值，margins = TRUE进行汇总
pd.pivot_table(df,index=[u'主客场'],columns=[u'对手'],
               values=[u'得分'],aggfunc=[np.sum],fill_value=0,margins=True)
```

```python
table = pd.pivot_table(df,index=[u'对手',u'胜负'],columns=[u'主客场'],
                       values=[u'得分',u'助攻',u'篮板'],aggfunc=[np.mean],
                      fill_value=0)
```
