```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('Mall_Customers.csv')
```

```python
data.head()
```

### EDA

```python
data.describe()
```

```python
sns.countplot(x = 'Gender' , data = data)
plt.title("年龄分布")
```

```python
data.hist("Age",bins = 35)
plt.title("年龄分布")
plt.xlabel("Age")
```

```python
# 按性别划分年龄直方图
plt.hist("Age",data = data[data["Gender"] == 'Male'],alpha = 0.5 , label = 'Male')
plt.hist("Age",data = data[data["Gender"] == "Female"],alpha = 0.5,label = "Female")
plt.title("按性别年龄分布")
plt.xlabel('Age')
plt.legend()
```

```python
# 收入柱图
data.hist("Annual Income (k$)")
plt.title("收入分布")
plt.xlabel("千$")
```

```python
# 收入按年龄
plt.hist("Annual Income (k$)",data = data[data["Gender"] == "Male"],alpha = 0.5,label = "Male")
plt.hist("Annual Income (k$)",data = data[data["Gender"] == "Female"], alpha = 0.5,label = "Female")
plt.title("收入按年龄分布")
plt.xlabel("美元收入")
plt.legend()
```

### 按性别创建数据集

```python
data_male = data[data["Gender"] == "Male"]
data_female = data[data["Gender"] == "Female"]
```

```python
# 计算性别花费均值
print(data_male['Spending Score (1-100)'].mean())
print(data_female['Spending Score (1-100)'].mean())
```

```python
sns.scatterplot("Age","Annual Income (k$)",hue = "Gender",data = data)
plt.title("hehe")
```

```python
sns.heatmap(data.corr(), annot=True);
plt.title('Heatmap Correlation of All Variables');
```

```python
sns.scatterplot('Age', 'Spending Score (1-100)', hue='Gender', data=data);
plt.title('Age to Spending Score, Colored by Gender');
```

```python
sns.heatmap(data_female.corr(), annot=True);
plt.title('Correlation Heatmap - Female');
```

```python
sns.heatmap(data_male.corr(), annot=True);
plt.title('Correlation Heatmap - Male');
```

```python
sns.lmplot('Age', 'Spending Score (1-100)', data=data_female);
plt.title('Age to Spending Score, Female Only');
```

```python
sns.scatterplot('Annual Income (k$)', 'Spending Score (1-100)', hue='Gender', data=data);
plt.title('Annual Income to Spending Score, Colored by Gender');
```
