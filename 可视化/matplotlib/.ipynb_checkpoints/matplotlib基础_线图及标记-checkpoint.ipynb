### 使用关键字字符串绘图

```python
data = {'a': np.arange(50),
        'c': np.random.randint(0, 50, 50),
        'd': np.random.randn(50)}
data['b'] = data['a'] + 10 * np.random.randn(50)
data['d'] = np.abs(data['d']) * 100

plt.scatter('a', 
            'b', 
            c='c', # 颜色 
            s='d', # 大小
            data=data)
# plt.scatter('a', 'b',s='d',data=data)

plt.xlabel('entry a')
plt.ylabel('entry b')
plt.show()
```

#### 使用多个图形和轴 

```python
def f(t):
    return np.exp(-t) * np.cos(2*np.pi*t)

t1 = np.arange(0.0, 5.0, 0.1)
t2 = np.arange(0.0, 5.0, 0.02)
```

```python
plt.figure(1)

plt.subplot(211)
plt.plot(t1, f(t1), 'bo', 
         t2, f(t2), 'k')

plt.subplot(212)
plt.plot(t2, np.cos(2*np.pi*t2), 'r--')

plt.show()
```

```python
import matplotlib.pyplot as plt
plt.figure(1)                # the first figure
plt.subplot(211)             # the first subplot in the first figure
plt.plot([1, 2, 3])

plt.subplot(212)             # the second subplot in the first figure
plt.plot([4, 5, 6])

plt.figure(2)                # a second figure
plt.plot([4, 5, 6])          # creates a subplot(111) by default

plt.figure(1)                # figure 1 current; subplot(212) still current
plt.subplot(211)             # make subplot(211) in figure1 current

plt.title('Easy as 1, 2, 3') # subplot 211 title
```

##### 设置坐标轴

```python
%matplotlib notebook
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(-3,3,50)
y1 = 2*x + 1
y2 = x**2
new_ticks = np.linspace(-1,2,5)
print(new_ticks)
```

```python
plt.figure(num = 6, # figure 6 
          figsize = (8,6)) # 图片大小
l1, = plt.plot(x,y1,label = "up") # 原始设置 # 想传到handles里就必须加逗号
l2, = plt.plot(x,  # 横轴数据  #想传到handles里就必须加逗号
         y2, # 纵轴数据
         label = "down",
        color = "red",   # 线颜色
        linewidth = 3,   # 线宽度
        linestyle = "--" # 线类型
        ) 
plt.legend(handles=[l1,l2,], # ？？？
           labels=['aaa','bbb'], # 这里设置标签就优先显示
           loc=1, # 图例位置
           fontsize = 20, # 字体大小
           ncol = 2, # 分两列
           title = "CCC" # 图例标题
          )


plt.xlim((-1,2)) # 修改x坐标轴取值范围
plt.ylim((-2,3)) # 修改y坐标轴取值范围
plt.xlabel("x轴") # 命名x轴
plt.ylabel("y轴") # 命名y轴
plt.xticks(new_ticks) # 修改x轴内容
plt.yticks([-2,-1.8,-1,1.22,3],[r"$A\ good$",r"$B\ \alpha$",r"$C$",r"$D$",r"$E$"])
# gca = "get current axis"
ax = plt.gca()
ax.spines["right"].set_color('none') # 隐藏右侧边框
ax.spines["top"].set_color('none')   # 隐藏上部边框
ax.xaxis.set_ticks_position("bottom") # 指定哪条轴为x轴
ax.yaxis.set_ticks_position("left")   # 指定哪条轴为y轴
# ax.spines["bottom"].set_position(("data",-1)) # ourward,axes(百分数)
ax.spines["bottom"].set_position(("axes",0.5))
ax.spines["left"].set_position(("data",0))
plt.show()
```

#### 图例位置：plt.legend(loc='lower left')

7: ‘center right'<br>

#### 标注

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(-3,3,50)
y = 2*x + 1
```

```python
plt.figure(num=1,figsize=(8,5))
plt.plot(x,y)

ax = plt.gca()
ax.spines['right'].set_color("none")
ax.spines["top"].set_color("none")
ax.xaxis.set_ticks_position("bottom")
ax.spines["bottom"].set_position(("data",0))
ax.yaxis.set_ticks_position("left")
ax.spines["left"].set_position(("data",0))

x0 = 1
y0 =2*x0 +1
plt.scatter(x0,y0,s = 50 , color = "red")
plt.plot([x0,x0],[y0,0],"k--",lw=2.5)
# 注释方法一
plt.annotate(r"$2x+1=%s$" % y0, # 注释内容
            xy = (x0,y0), # 从哪开始打印
            xycoords = 'data', # 基准是data
            xytext = (+30,-30),
            textcoords = 'offset points', # 从原点在移动右30，下30开始打印
            fontsize = 16,
            arrowprops = dict(arrowstyle = "->",connectionstyle = "arc3,rad=.5") # 箭头类型
            ) 

# 注释方法二
plt.text(-3.7,3,r"$This\ is\ the\ some\ text.\ \mu\ \sigma_i\ \alpha_t$",
        fontdict = {"size":16,'color':'r'}
        )

plt.show()

```

### 坐标轴设置

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(-3,3,50)
y = 0.1*x
```

```python
plt.figure(num=1,figsize=(8,5))
plt.plot(x,y,linewidth = 10,alpha = 0.7,color = "g")
plt.ylim(-2,2)

ax = plt.gca()
ax.spines['right'].set_color("none")
ax.spines["top"].set_color("none")
ax.xaxis.set_ticks_position("bottom")
ax.spines["bottom"].set_position(("data",0))
ax.yaxis.set_ticks_position("left")
ax.spines["left"].set_position(("data",0))

for label in ax.get_xticklabels() + ax.get_yticklabels():
    label.set_fontsize(12)
    label.set_bbox(dict(facecolor = "white",
                       edgecolor = "None",
                        alpha = 1.5
                       ))

plt.show()
```

### 多图(网格)

```python
import matplotlib.pyplot as plt 
import matplotlib.gridspec as gridspec
```

```python
# 方法1：subplot2grid
plt.figure(figsize = (10,8))
ax1 = plt.subplot2grid((3,3), #3行3列
                       (0,0), # 从0,0开始
                       colspan = 3, # 占3列
                       rowspan = 1, # 占1行
                      )
ax1.plot([1,2],[1,2])
ax1.set_title("ax1_title")
ax2 = plt.subplot2grid((3,3),(1,0),colspan = 2,rowspan = 1)
ax3 = plt.subplot2grid((3,3),(1,2),colspan = 1,rowspan = 2)
ax4 = plt.subplot2grid((3,3),(2,0),colspan = 1,rowspan = 1)
ax5 = plt.subplot2grid((3,3),(2,1),colspan = 1,rowspan = 1)
plt.tight_layout()
plt.show()
```

```python
# 方法二
plt.figure(figsize = (8,6))
gs = gridspec.GridSpec(3,3) # 3行3列
ax1 = plt.subplot(gs[0,:3])
ax1.set_title("1")
ax2 = plt.subplot(gs[1,:2])
ax2.set_title("2")
ax3 = plt.subplot(gs[1:,2])
ax3.set_title("3")
ax4 = plt.subplot(gs[-1,0])
ax4.set_title("4")
ax5 = plt.subplot(gs[-1,-2])
ax5.set_title("5")
plt.tight_layout()
plt.show()
```

```python
# 方法三
f,((ax11,ax12),(ax21,ax22)) = plt.subplots(2,2, # 2行2列
             sharex = True, # 共享x轴
             sharey = True, # 共享y轴
            )
ax11.scatter([1,2],[1,2])
plt.tight_layout()
plt.show()
```
