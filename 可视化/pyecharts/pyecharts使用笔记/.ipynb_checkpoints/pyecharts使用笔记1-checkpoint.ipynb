```python
import pyecharts
from pyecharts import Bar, Line
from pyecharts.engine import create_default_environment


import random
```

```python
#定义标题
bar = Bar("我的第一个图表", "这里是副标题")

#使用主题(echarts-themes-pypkg)
bar.use_theme('dark') #默认dark

#主要方法，用于添加图表的数据和设置各种配置项
bar.add("服装",
        ["衬衫", "羊毛衫", "雪纺衫", "裤子", "高跟鞋", "袜子"],
        [5, 20, 36, 10, 75, 90],
        is_more_utils=True)
# #打印输出图表的所有配置项
# #bar.print_echarts_options() # 该行只为了打印配置项，方便调试时使用

# 默认将会在根目录下生成一个 render.html 的文件，
# 支持 path 参数，设置文件保存位置，如 render(r"e:\my_first_chart.html")，文件用浏览器打开。
# bar.render()    # 生成本地 HTML 文件
```


## 使用 pyecharts-snapshot 插件

<h2>如果想直接将图片保存为 png, pdf, gif 格式的文件，可以使用 pyecharts-snapshot。使用该插件请确保你的系统上已经安装了 Nodejs 环境。</h2>

安装 phantomjs：npm install -g phantomjs-prebuilt<br>
安装 pyecharts-snapshot：pip install pyecharts-snapshot<br>
调用 render 方法 bar.render(path='snapshot.png') 文件结尾可以为 svg/jpeg/png/pdf/gif。<br>请注意，svg 文件需要你在初始化 bar 的时候设置 renderer='svg'。

## 图形绘制过程
1、实例一个具体类型图表的对象：	chart = FooChart()	
2、为图表添加通用的配置，如主题：	chart.use_theme()	
3、为图表添加特定的配置：	geo.add_coordinate()	
4、添加数据及配置项：	chart.add()	
5、生成本地文件（html/svg/jpeg/png/pdf/gif）：	chart.render()

##### 支持链式调用

```python
CLOTHES = ["衬衫", "羊毛衫", "雪纺衫", "裤子", "高跟鞋", "袜子"]
clothes_v1 = [5, 20, 36, 10, 75, 90]
clothes_v2 = [10, 25, 8, 60, 20, 80]

(Bar("柱状图数据堆叠示例")
    .add("商家A", CLOTHES, clothes_v1, is_stack=True)
    .add("商家B", CLOTHES, clothes_v2, is_stack=True)
    .render())
```

```python


bar = Bar("我的第一个图表", "这里是副标题")
bar.add("服装", ["衬衫", "羊毛衫", "雪纺衫", "裤子", "高跟鞋", "袜子"], [5, 20, 36, 10, 75, 90])

line = Line("我的第一个图表", "这里是副标题")
line.add("服装", ["衬衫", "羊毛衫", "雪纺衫", "裤子", "高跟鞋", "袜子"], [5, 20, 36, 10, 75, 90])

env = create_default_environment("html")
# 为渲染创建一个默认配置环境
# create_default_environment(filet_ype)
# file_type: 'html', 'svg', 'png', 'jpeg', 'gif' or 'pdf'

env.render_chart_to_file(bar, path='bar.html')
env.render_chart_to_file(line, path='line.html')
```
