```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style="darkgrid")
```

```python
tips = sns.load_dataset("tips")
```

```python
tips.head()
```

```python
sns.relplot(x="total_bill", 
            y="tip", 
            hue="smoker", # 色调
            col="time",  # 分图
            data=tips); # 数据源
```

```python
fmri = sns.load_dataset("fmri")
fmri.head()
```

```python
sns.relplot(x="timepoint", 
            y="signal", 
            hue="subject",
            col="region", 
            row="event", 
            height=5, # 图大小
            kind="line", 
            estimator=None, 
            data=fmri);
```
