```python
import matplotlib.pyplot as plt
from matplotlib.font_manager import FontProperties

chinese = FontProperties(fname = 'C:/Windows/Fonts/msyh.ttc')
plt.rcParams['font.sans-serif'] = [u'SimHei']
plt.rcParams['axes.unicode_minus'] = False
```

```python
plt.plot([1, 2, 3, 4])
plt.ylabel('y轴标签')
plt.xlabel('x轴标签')
plt.show()
```

```python
# 要绘制x与y的关系
plt.plot([1, 2, 3, 4], [1, 4, 9, 16])
```

### 线样式格式：fmt = '[marker][line][color]'

```python
# 点样式
plt.plot([1, 2, 3, 4], [1, 4, 9, 16], '-.') # 第三个参数，颜色和线型
plt.axis([0, 6, 0, 20]) # x,y轴距
plt.show()
```

```python
# 线样式
plt.plot([1, 2, 3, 4], [1, 4, 9, 16], '-.') # 第三个参数，颜色和线型
plt.axis([0, 6, 0, 20]) # x,y轴距
plt.show()
```

##### 线形编码

'-'	solid line style 实线 <br>
'--'	dashed line style 大虚线<br>
'-.'	dash-dot line style 点划线 顺序不同样式不同<br>
':'	dotted line style 小虚线

```python
# 颜色样式
plt.plot([1, 2, 3, 4], [1, 4, 9, 16], '^g') # 第三个参数，颜色和线型
plt.axis([0, 6, 0, 20]) # x,y轴距
plt.show()
```

##### 颜色编码

'b'	blue<br>
'g'	green<br>
'r'	red<br>
'c'	cyan<br>
'm'	magenta<br>
'y'	yellow<br>
'k'	black<br>
'w'	white
