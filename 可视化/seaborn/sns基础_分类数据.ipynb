```python

%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib as mpl
import seaborn as sns
import matplotlib.pyplot as plt
sns.set(style="ticks", color_codes=True) # whitegrid
```

```python
tips = sns.load_dataset("tips")
tips.head()
```

```python
sns.stripplot(x="day",y="total_bill",data=tips)
```

```python
sns.catplot(x="day", 
            y="total_bill", 
            data=tips);
```

```python
sns.catplot(x="day", 
            y="total_bill", 
            jitter=False, # 排一排
            data=tips);
```

```python
sns.catplot(x="day",y="total_bill",kind="swarm",data=tips)
# sns.swarmplot(x="day",y="total_bill",data=tips)
```

```python
sns.catplot(x="day", 
            y="total_bill", 
            hue="sex", # 色调
            kind="swarm", 
            data=tips);
# sns.swarmplot(x="day",y="total_bill",hue="sex",data=tips)
```

```python
sns.catplot(x="size", 
            y="total_bill", 
            kind="swarm",
            data=tips.query("size != 3"));
```

```python
sns.catplot(x="smoker", 
            y="tip", 
            order=["No", "Yes"], 
            data=tips)
```

```python
sns.catplot(x="total_bill", 
            y="day", 
            hue="time", 
            kind="swarm", 
            data=tips);
```
