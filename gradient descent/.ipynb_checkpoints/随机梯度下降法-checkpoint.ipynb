```python
import pandas as pd
import numpy as np
```

```python
data_train = pd.read_csv('D:/backup/Machine Learning/gradient descent/data_questions/train.csv')
data_test = pd.read_csv('/backup/Machine Learning/gradient descent/data_questions/test.csv')
data_submit = pd.read_csv('/backup/Machine Learning/gradient descent/data_questions/test.csv')
```

```python
#初始设置
theta = [1,1]
alpha = 0.2
tol_l = 0.1
```

```python
#对x进行归一化
max_x = max(data_train['id'])
x = data_train['id'] / max_x
y = data_train['questions']
```

```python
#定义计算随机梯度的函数
def compute_SGD(theta,x,y):
    gradient = [0,0]
    r = np.random.randint(0,len(x))
    gradient[0] = 2. * np.mean( theta[0] + theta[1] * x[r] - y[r])
    gradient[1] = 2. * np.mean((theta[0] + theta[1] * x[r] - y[r])*x[r])
    return np.array(gradient)
```

```python
#定义theta函数
def update_theta(theta,alpha,gradient):
    theta_new = np.array(theta) - alpha * gradient
    return theta_new
```

```python
#定义计算RMSE的函数
def rmse(theta,x,y):
    squared_error = (theta[0] + theta[1] * x - y) **2
    result = np.sqrt(np.mean(squared_error))
    return result
```

```python
#进行第一次计算
np.random.seed(10)
gradient = compute_SGD(theta,x,y)
loss = rmse(theta,x,y)
theta = update_theta(theta,alpha,gradient)
loss_new = rmse(theta,x,y)
```

```python
#开始迭代
i = 1
while np.abs(loss_new - loss) > tol_l:
    theta = update_theta(theta,alpha,gradient)
    gradient = compute_SGD(theta,x,y)
    if i % 100 == 0:
        loss = loss_new
        loss_new = rmse(theta,x,y)
        print('第%s回，均方根误差：%s'%(i,abs(loss_new - loss)))
    i += 1
print('theta[1]：%s  \ntheta[0]：%s'%(theta[1],theta[0]))
```

```python
#最后参数(归一化回来)
print('最终theta[1]:%s \n最终theta[0]:%s'%(theta[1] / max_x , theta[0]))
```

```python
#训练误差RMSE
result = rmse(theta,x,y)
print('最终误差：%s'%result)
```
