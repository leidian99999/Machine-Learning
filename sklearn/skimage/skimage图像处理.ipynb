```python
import numpy as np
import scipy as sp
import matplotlib.pyplot as plt
from skimage import io
```


io	读取,保存和显示图片和视频 <br>
color	颜色空间变换<br>
data	提供一些测试图片和样本数据<br>
filters	图像增强，边缘检测，排序滤波器，自动阈值等<br>
draw	操作于numpy数组上的基本图形绘制，包括线条，矩阵，圆和文本等<br>
transform	几何变换和其他变换，如旋转，拉伸和Radon(拉东)变换等<br>
exposure	图像强度调整，例如，直方图均衡化等<br>
feature	特征检测和提取， 例如，纹理分析等<br>
graph	图论操作，例如，最短路径<br>
measure	图像属性测量，例如，相似度和轮廓<br>
morphology	形态学操作，如开闭运算，骨架提取等<br>
novice	简化的用于教学目的的接口<br>
restoration	修复算法，例如去卷积算法，去噪等<br>
segmentation	图像分割为多个区域<br>
util	通用工具<br>
viewer	简单图形用户界面用于可视化结果和探索参数<br>
